```python
%%time
#Transform

df=pd.DataFrame({})

for dem_col_name,product_id in enumerate(stv['id'][:10]):
    
    data=stv[stv.id==product_id].iloc[:,6:].T.reset_index()
    data=data.merge(cal, left_on='index', right_on='d')
    data=data.drop(['d','snap_CA', 'snap_WI','date'],axis=1)

    #Get the prices
    store_id='_'.join(product_id.split('_')[3:5])
    item_id='_'.join(product_id.split('_')[:3])
    df_price=sellp[sellp.item_id==item_id]
    df_price=df_price[(df_price.store_id==store_id)]
    df_price=df_price[df_price.wm_yr_wk.isin(wm_yr_wk_unique)]

    #Merge it
    data=data.merge(df_price, on='wm_yr_wk')
    df=df.append(data)
```
